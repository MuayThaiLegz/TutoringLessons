```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x * 0.1)

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1: Sine
axs[0, 0].plot(x, y1, 'tab:blue')
axs[0, 0].set_title('Sine Function')

# Plot 2: Cosine
axs[0, 1].plot(x, y2, 'tab:orange')
axs[0, 1].set_title('Cosine Function')

# Plot 3: Tangent
axs[1, 0].plot(x, y3, 'tab:green')
axs[1, 0].set_title('Tangent Function')
axs[1, 0].set_ylim(-10, 10)  # Limit y-axis to avoid extreme values

# Plot 4: Exponential
axs[1, 1].plot(x, y4, 'tab:red')
axs[1, 1].set_title('Exponential Function')

# Automatically adjust subplot params for better layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for the sine wave
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Sample some points to highlight
sample_x = np.linspace(0, 10, 15)
sample_y = np.sin(sample_x)

# Create the line plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave', color='blue')

# Overlay with a scatter plot
plt.scatter(sample_x, sample_y, color='red', label='Sampled Points')

# Adding customizations
plt.title('Overlaying Line and Scatter Plots')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the built-in dataset 'naturalearth_lowres' in Geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Suppose we have a hypothetical GDP per capita data (in USD)
# This is just random data for demonstration purposes
import numpy as np
world['gdp_per_cap'] = np.random.randint(500, 80000, size=len(world))

# Plotting
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.plot(column='gdp_per_cap', ax=ax, legend=True,
           legend_kwds={'label': "GDP per Capita (USD)", 'orientation': "horizontal"},
           cmap='OrRd')  # Using a colormap for GDP per capita

# Customizations
ax.set_title('World Countries by GDP per Capita', fontsize=15)
ax.set_axis_off()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Set the Seaborn style
sns.set_style('whitegrid')  # Other styles include 'darkgrid', 'white', 'dark', 'ticks'

# Define a custom color palette
custom_palette = ['#4B0082', '#FFD700', '#00FF00', '#FF69B4']  # Indigo, Gold, Lime, Hot Pink
sns.set_palette(custom_palette)

# Create a scatter plot
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, s=100, alpha=0.7, edgecolor='w')

# Customizing the plot
scatter_plot.set_title('Iris Sepal Dimensions', fontsize=18)
scatter_plot.set_xlabel('Sepal Length (cm)', fontsize=12)
scatter_plot.set_ylabel('Sepal Width (cm)', fontsize=12)
scatter_plot.legend(title='Species', title_fontsize='13', labelspacing=1.05)

# Show the plot
plt.show()
```
