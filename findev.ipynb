```python
datasets = [income_df, balance_df, cashflow_df]

[j.reset_index(inplace= True) for j in datasets]
[i.set_index(['Ticker', 'Publish Date'], inplace= True) for i in datasets]

[i.drop(columns = ['Restated Date','SimFinId', 'Currency','Fiscal Year','Fiscal Period','index','Report Date'],
        inplace= True, errors='ignore') for i in datasets]

[data.fillna(data.mean(), inplace=True) for data in datasets]

datasetsnoshares = [income_df, balance_df]

[i.drop(columns = ['Shares (Diluted)','Shares (Basic)'], inplace= True, errors='ignore') for i in datasetsnoshares]

```

```python

skldata = []

[skldata.append((data - data.mean())/data.std()) for data in datasets]
```

```python
skl_income_df = skldata[0]
skl_balance_df = skldata[1]
skl_cashflow_df = skldata[2]
```
